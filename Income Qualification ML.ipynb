```python
obj_variable=list(train.select_dtypes(include=['object']).columns)
```

```python
obj_variable
```

```python
#Label Encoding for object to numeric conversion
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feat in obj_variable:
    train[feat] = le.fit_transform(train[feat].astype(str))

print (train.info())
```

```python
train.head()
```

```python
for feat in obj_variable:
    test[feat] = le.fit_transform(test[feat].astype(str))

print (test.info())
```

```python
X=train.iloc[:,:-1].values
y=train.iloc[:,-1].values
```

```python
# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    random_state=1)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=600, max_depth=300, max_features='sqrt')
rfc.fit(X_train,y_train)
rfc_predict = rfc.predict(X_test)
scores = cross_val_score(rfc, X_train, y_train, cv=5)
print(scores)
print(metrics.accuracy_score(rfc_predict,y_test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, rfc_predict))
print('\n')
print("=== Classification Report ===")
print(classification_report(y_test, rfc_predict))
```
